```python
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
    def __init__(self, num_inputs):
        self.weights = np.zeros(num_inputs + 1)  # +1 for the bias term

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return 1 if summation > 0 else 0

    def train(self, inputs, targets, num_epochs):
        for _ in range(num_epochs):
            for input_data, target in zip(inputs, targets):
                prediction = self.predict(input_data)
                error = target - prediction
                self.weights[1:] += error * input_data
                self.weights[0] += error

def visualize_data(X, Y):
    area = 200
    fig = plt.figure(figsize=(6, 6))
    plt.title('The OR Gate', fontsize=20)
    ax = fig.add_subplot(111)
    for x, y in zip(X, Y):
        color = 'r' if y == 0 else 'b'
        ax.scatter(x[0], x[1], s=area, c=color, label="Class {}".format(y))
    plt.grid()
    plt.show()

# Data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 1])

# Create and train the perceptron
perceptron = Perceptron(num_inputs=2)
perceptron.train(X, Y, num_epochs=10)

# Visualize the OR gate
visualize_data(X, Y)

# Test the trained perceptron
test_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
for data in test_data:
    prediction = perceptron.predict(data)
    print("Input: {}, Prediction: {}".format(data, prediction))
```
